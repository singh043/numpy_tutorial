## Numpy Arrays

### Basic Array Creation

```python
arr = np.array([1, 2, 3, 4, 5]) # Create a NumPy array from a list
print("Array:", arr)
a = [1, 2, 3, "j"] # Create a list with mixed data types
arr1 = np.array(a)
print(arr1)
```

### Array generation function

```python
arr = np.arange(1, 11) # Create a NumPy array with values from 1 to 10 (exclusive)
print("Array:", arr)

arr1 = np.arange(1, 11, 2) # Create a NumPy array with values from 1 to 10 (exclusive) and a step of 2
print("Array with step 2:", arr1)
```

```python
arr = np.zeros(6)
print("Array of zeros:", arr)

arr = np.zeros((2, 3))
print("2D Array of zeros:\n", arr)
```

```python
arr = np.ones(5) # Create a NumPy array of ones with length 5
print("Array with ones:\n", arr)

arr = np.ones((3, 4)) # Create a 2D NumPy array of ones with shape (3, 4)
print("2D Array with ones:\n", arr)
```

```python
arr = np.linspace(1, 20, 5) # Create a NumPy array with 5 linearly spaced values between 1 and 20 (inclusive)
print("Linearly spaced array:", arr)
```

### Random Generation Function

### Deep & Shallow Copy

### shallow copy

```python
a = np.arange(1, 21)
a
```

```python
slice_a = a[:5]
slice_a
```

```python
slice_a * 10 # This modifies the slice, but not the original array a
```

```python
a
```

```python
a[4:6] * 10 # This does not modify the original array a
```

```python
b = a.copy() # This creates a new array that is a copy of a
b
```

```python
b[0] = 100
print("Array a:", a)
print("Array b:", b)
```

### deep copy

```python
b = a # This creates a reference, not a copy
a
```

```python
b[0] = 100
print("Array a:", a)
print("Array b:", b)
```

### Valid Sudoku

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
s1 = np.sum(s, axis = 0)

for x in s1:
    if x != 45:
        print("Not a valid sudoku")
        r = 1
        break

s2 = np.sum(s, axis = 1)

for x in s2:
    if x != 45:
        print("Not a valid sudoku")
        r = 1
        break


for i in range(0,9,3):
    for j in range(0,9,3):
        np.sum(s[i:i+3,j:j+3]) != 45
        r = 1
        break
    
if r != 0:
    print("Valid sudoku")
```

```python
import numpy as np

# Columns: [Age, Math Marks, Science Marks]

data = np.array([
    [18, 85, 78], # Student 1
    [19, 92, 88], # Student 2
    [17, 76, 95], # Student 3
    [18, 65, 70], # Student 4
    [20, 90, 85]  # Student 5
])
```

```python
# Get the shape of the matrix.

data.shape
```

```python
# Find the average age of the students.
arr = data[:,0]
print("Ages of students:", arr)
average_age = arr.mean()
print("Average age of students:", average_age)
```

```python
# Extract Math marks of all students.
data [:,1]
```

```python
# Find the highest Science marks.

np.max(data[:,2])
```

```python
# Get details of students who scored more than 90 in Math.
data[data[:,1] > 90]
```

```python
# Increase Math marks of all students by 5.

data[:,1] = data[:,1] + 5
data
```

```python
# Find how many students are younger than 19.

len(data[data[:,0] < 19])
```

```python
# Calculate the average marks in each subject (Column-wise).
Math_avg = data[:,1].mean()
Science_avg = data[:,2].mean()
print("Average Math marks:", Math_avg)
print("Average Science marks:", Science_avg)
```

```python
# OR

np.mean(data[:,1:], axis = 0)
```

```python
# Get data of students who scored atleast 80 in both Math and Science.

data[(data[:,1] >= 80) & (data[:,2] >= 80)][:,0]
```

```python
# Replace all Science marks < 75 with 0.

data[:,2][data[:,2] < 75] = 0
data
```

```python
# OR

data[:,2] = [0 if data[x,2] < 75 else data[x,2] for x in range(0, len(data)) ]
print(data)
```

```python
# OR

for x in data:
    if x[2] < 75:
        x[2] = 0
print(data)
```
